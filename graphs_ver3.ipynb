```python
import KEGGutils as kg
import networkx as nx

mcc_glyco = kg.KEGGpathway(pathway_id = "mcc00010")
# print(mcc_glyco.title)
mcc_glyco.draw()

mcc_tca = kg.KEGGpathway(pathway_id = "mcc00020")
# print(mcc_tca.title)
mcc_tca.draw()
```

```python
gtt_glyco = kg.KEGGpathway(pathway_id = "gtt00010")
# print(gtt_glyco.title)
gtt_glyco.draw()

gtt_tca = kg.KEGGpathway(pathway_id = "gtt00020")
# print(gtt_tca.title)
gtt_tca.draw()
```

```python
ffo_glyco = kg.KEGGpathway(pathway_id = "ffo00010")
# print(ffo_glyco.title)
ffo_glyco.draw()

ffo_tca = kg.KEGGpathway(pathway_id = "ffo00020")
# print(ffo_tca.title)
ffo_tca.draw()
```

```python
lrm_glyco = kg.KEGGpathway(pathway_id = "lrm00010")
# print(ffo_glyco.title)
lrm_glyco.draw()

lrm_tca = kg.KEGGpathway(pathway_id = "lrm00020")
# print(ffo_tca.title)
lrm_tca.draw()
```

```python
ngr_glyco = kg.KEGGpathway(pathway_id = "ngr00010")
# print(ffo_glyco.title)
ngr_glyco.draw()

ngr_tca = kg.KEGGpathway(pathway_id = "ngr00020")
# print(ffo_tca.title)
ngr_tca.draw()
```

```python
mcc_glyco_edges = list(dict.fromkeys(mcc_glyco.edges))
# print("MCC GLYCO EDGES: ",mcc_glyco_edges, "\n\n")
print("MCC GLYCO EDGES: ", len(mcc_glyco_edges))

gtt_glyco_edges= list(dict.fromkeys(gtt_glyco.edges))
# print("GTT GLYCO EDGES: ",gtt_glyco_edges, "\n\n")
print("GTT GLYCO EDGES: ", len(gtt_glyco_edges))

ffo_glyco_edges= list(dict.fromkeys(ffo_glyco.edges))
# print("FFO GLYCO EDGES: ",ffo_glyco_edges, "\n\n")
print("FFO GLYCO EDGES: ", len(ffo_glyco_edges))

lrm_glyco_edges= list(dict.fromkeys(lrm_glyco.edges))
# print("LRM GLYCO EDGES: ",lrm_glyco_edges, "\n\n")
print("LRM GLYCO EDGES: ", len(lrm_glyco_edges))

ngr_glyco_edges= list(dict.fromkeys(ngr_glyco.edges))
# print("NGR GLYCO EDGES: ",ngr_glyco_edges)
print("NGR GLYCO EDGES: ", len(ngr_glyco_edges))
```

```python
all_glyco_edges = []
all_glyco_edges.extend(mcc_glyco_edges)
all_glyco_edges.extend(gtt_glyco_edges)
all_glyco_edges.extend(ffo_glyco_edges)
all_glyco_edges.extend(lrm_glyco_edges)
all_glyco_edges.extend(ngr_glyco_edges)
# print(all_glyco_edges,"\n\n")
print("ALL GLYCO EDGES: ",len(all_glyco_edges),"\n")

all_glyco_edges= list(dict.fromkeys(all_glyco_edges))
# print(all_glyco_edges,"\n\n")
print("UNIQUE GLYCO EDGES: ",len(all_glyco_edges),"\n")
```

```python
def setBit(bit, arr, edge):
    if edge in arr:
        bit += "1"
    else:
        bit += "0"
    return bit

ft_glyco = {}        
        
for edge in all_glyco_edges:
    bit = ""
    bit = setBit(bit,mcc_glyco_edges,edge)
    bit = setBit(bit,gtt_glyco_edges,edge)
    bit = setBit(bit,ffo_glyco_edges,edge)
    bit = setBit(bit,lrm_glyco_edges,edge)
    bit = setBit(bit,ngr_glyco_edges,edge)
    
    ft_glyco[edge] = bit

print("Frequency Table (Glycolysis) \n")
for key, value in ft_glyco.items():
    print(key, ' : ', value)
        
```

```python
mcc_tca_edges = list(dict.fromkeys(mcc_tca.edges))
# print("MCC TCA EDGES: ",mcc_tca_edges, "\n\n")
print("MCC TCA EDGES: ", len(mcc_tca_edges))

gtt_tca_edges= list(dict.fromkeys(gtt_tca.edges))
# print("GTT TCA EDGES: ",gtt_tca_edges, "\n\n")
print("GTT TCA EDGES: ", len(gtt_tca_edges))

ffo_tca_edges= list(dict.fromkeys(ffo_tca.edges))
# print("FFO TCA EDGES: ",ffo_tca_edges, "\n\n")
print("FFO TCA EDGES: ", len(ffo_tca_edges))

lrm_tca_edges= list(dict.fromkeys(lrm_tca.edges))
# print("LRM TCA EDGES: ",lrm_tca_edges, "\n\n")
print("LRM TCA EDGES: ", len(lrm_tca_edges))

ngr_tca_edges= list(dict.fromkeys(ngr_tca.edges))
# print("NGR TCA EDGES: ",ngr_tca_edges)
print("NGR TCA EDGES: ", len(ngr_tca_edges))
```

```python
all_tca_edges = []
all_tca_edges.extend(mcc_tca_edges)
all_tca_edges.extend(gtt_tca_edges)
all_tca_edges.extend(ffo_tca_edges)
all_tca_edges.extend(lrm_tca_edges)
all_tca_edges.extend(ngr_tca_edges)
# print(all_tca_edges,"\n\n")
print("ALL TCA EDGES: ",len(all_tca_edges),"\n")

all_tca_edges= list(dict.fromkeys(all_tca_edges))
# print(all_tca_edges,"\n\n")
print("UNIQUE TCA EDGES: ",len(all_tca_edges),"\n")
```

```python
ft_tca = {}
for edge in all_tca_edges:
    bit = ""
    bit = setBit(bit,mcc_tca_edges,edge)
    bit = setBit(bit,gtt_tca_edges,edge)
    bit = setBit(bit,ffo_tca_edges,edge)
    bit = setBit(bit,lrm_tca_edges,edge)
    bit = setBit(bit,ngr_tca_edges,edge)
    
    ft_tca[edge] = bit

print("Frequency Table (Citrate Cycle) \n")
for key, value in ft_tca.items():
    print(key, ' : ', value)
print("\n")
```

```python
# to create clusters
from collections import defaultdict    
    
ft_tca_cluster = defaultdict(list)
for key, value in sorted(ft_tca.items()):
    ft_tca_cluster[value].append(key)
for key, value in ft_tca_cluster.items():
    print(key, ' : ', value,"\n")
```
