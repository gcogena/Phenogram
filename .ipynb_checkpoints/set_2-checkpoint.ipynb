```python
import KEGGutils as kg
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster import hierarchy
from collections import defaultdict 
```

```python
# to set bitCode per edge
def setBit(bit, arr, edge):
    if edge in arr:
        bit += "1"
    else:
        bit += "0"
    return bit

# to get cardinality of intersection of 2 graphs
def getIntersection(edges1, edges2):
    return len(list(set(edges1) & set(edges2)))

# to get cardinality of union of 2 graphs
def getUnion(edges1, edges2,intersection):
    return (len(edges1) + len(edges2)) - intersection
```

```python
################ GLYCOLYSIS #####################
```

```python
# to graph glycolysis pathway

aqu_glyco = kg.KEGGpathway(pathway_id = "aqu00010")
# print(aqu_glyco.title)
aqu_glyco.draw()

gla_glyco = kg.KEGGpathway(pathway_id = "gla00010")
# print(gla_glyco.title)
gla_glyco.draw()

vcn_glyco = kg.KEGGpathway(pathway_id = "vcn00010")
# print(vcn_glyco.title)
vcn_glyco.draw()

hav_glyco = kg.KEGGpathway(pathway_id = "hav00010")
# print(hav_glyco.title)
hav_glyco.draw()

rtp_glyco = kg.KEGGpathway(pathway_id = "rtp00010")
# print(rtp_glyco.title)
rtp_glyco.draw()
```

```python
# to get distinct edges

aqu_glyco_edges = list(dict.fromkeys(aqu_glyco.edges))
# print("aqu GLYCO EDGES: ",aqu_glyco_edges, "\n\n")
print("aqu GLYCO EDGES: ", len(aqu_glyco_edges))

gla_glyco_edges= list(dict.fromkeys(gla_glyco.edges))
# print("gla GLYCO EDGES: ",gla_glyco_edges, "\n\n")
print("gla GLYCO EDGES: ", len(gla_glyco_edges))

vcn_glyco_edges= list(dict.fromkeys(vcn_glyco.edges))
# print("vcn GLYCO EDGES: ",vcn_glyco_edges, "\n\n")
print("vcn GLYCO EDGES: ", len(vcn_glyco_edges))

hav_glyco_edges= list(dict.fromkeys(hav_glyco.edges))
# print("hav GLYCO EDGES: ",hav_glyco_edges, "\n\n")
print("hav GLYCO EDGES: ", len(hav_glyco_edges))

rtp_glyco_edges= list(dict.fromkeys(rtp_glyco.edges))
# print("rtp GLYCO EDGES: ",rtp_glyco_edges)
print("rtp GLYCO EDGES: ", len(rtp_glyco_edges))

all_glyco_edges = []
all_glyco_edges.extend(aqu_glyco_edges)
all_glyco_edges.extend(gla_glyco_edges)
all_glyco_edges.extend(vcn_glyco_edges)
all_glyco_edges.extend(hav_glyco_edges)
all_glyco_edges.extend(rtp_glyco_edges)
# print(all_glyco_edges,"\n\n")
print("\n ALL GLYCO EDGES: ",len(all_glyco_edges))

all_glyco_edges= list(dict.fromkeys(all_glyco_edges))
# print(all_glyco_edges,"\n\n")
print("UNIQUE GLYCO EDGES: ",len(all_glyco_edges))
```

```python
# to create frequency table

ft_glyco = {}        
        
for edge in all_glyco_edges:
    bit = ""
    bit = setBit(bit,aqu_glyco_edges,edge)
    bit = setBit(bit,gla_glyco_edges,edge)
    bit = setBit(bit,vcn_glyco_edges,edge)
    bit = setBit(bit,hav_glyco_edges,edge)
    bit = setBit(bit,rtp_glyco_edges,edge)
    
    ft_glyco[edge] = bit

print("Frequency Table (Glycolysis) \n")
for key, value in ft_glyco.items():
    print(key, ' : ', value)
```

```python
# to create nodes of glycolysis freq. table 
    
ft_glyco_nodes = defaultdict(list)
for key, value in sorted(ft_glyco.items()):
    ft_glyco_nodes[value].append(key)
for key, value in ft_glyco_nodes.items():
    print(key, ' : ', value,"\n")
```

```python
# to create clusters of glycolysis freq. table 

ft_glyco_cluster = defaultdict(list)

for key in ft_glyco_nodes.keys():
    count = str(key).count("1")
    ft_glyco_cluster[count].append(key)

for key, value in sorted(ft_glyco_cluster.items()):
    print(key, ' : ', value,"\n")
```

```python
glyco_dict = defaultdict(list)
glyco_dict[0] = aqu_glyco_edges
glyco_dict[1] = gla_glyco_edges
glyco_dict[2] = vcn_glyco_edges
glyco_dict[3] = hav_glyco_edges
glyco_dict[4] = rtp_glyco_edges

# to create jaccard similarity index for glycolysis
glyco_jsi = [[0 for c in range(5)] for r in range(5)]

for x in range(0,5):
    for y in range(0,5):
        i = getIntersection(glyco_dict[x],glyco_dict[y])
        u = getUnion(glyco_dict[x],glyco_dict[y],i)
        
        if i/u == 1:
            glyco_jsi[x][y] = 0
        else:
            glyco_jsi[x][y] = i/u

for x in glyco_jsi:
    print(x)
```

```python
df = pd.DataFrame.from_dict(glyco_jsi,orient='columns')
df.columns=['aqu', 'gla', 'vcn', 'hav', 'rtp']
df.index=['aqu', 'gla', 'vcn', 'hav', 'rtp']
df[:] = np.tril(df.values, k=-1)
print(df)
```
